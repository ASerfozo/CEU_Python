### API - HDRO data

### Get data and create a wide format dataset
Indicators for 2016 and 2019

Using:
- 148306: Employment to population ratio (% ages 15 and older)
- 194906: GDP per capita
- 137506: Human Development Index (HDI)
- 31706 : Share of seats in parliament (% held by women)

Decided not to use
- 53506 : Foreign direct investment, net inflows (% of GDP) --- 2/3 of the received 2019 data was NA
- 101406: Literacy rate, adult (% ages 15 and older) --- data only for 2016, with many NA
- 128306: Prison population (per 100,000 people) --- data only for 2016, with many NA
- 194306: Share of seats held by women in local government (%) --- no data for 2016 and 2019
- 179406: Skilled labour force (% of labour force) --- 2/3 of the received data was NA
- 153706: Working poor at PPP$3.20 a day (% of total employment) --- 1/3 of the received data was NA

```python
# Get request from Human Development Report Office Statistical Data API
resp = requests.get("http://ec2-54-174-131-205.compute-1.amazonaws.com/API/HDRO_API.php/indicator_id=148306,137506,194906,31706/year=2016,2019")
```

```python
# Load response into json
resp = json.loads(resp.text)
resp
```

```python
# Get all data in long format
df_rows = []    
for c in resp['indicator_value'].keys():
    #print('country', c)
    for i in resp['indicator_value'][c].keys():
        #print('indicator', i)
        for y in resp['indicator_value'][c][i].keys():
            #print('year', y)
            print([c,resp['indicator_value'][c][i][y]])
            df_rows.append([c,i,y,resp['indicator_value'][c][i][y]])
df_rows
```

```python
# Convert to dataframe
long_df = pd.DataFrame(df_rows, columns=['country_code','indicator','year','value'])
long_df
```

```python
# Rename the indicator numbers to be easier to understand
long_df['indicator'] = np.where(long_df['indicator'] == "148306", "Employment_ratio" , 
                       np.where(long_df['indicator'] == "194906", "GDP_per_capita" ,
                       np.where(long_df['indicator'] == "137506", "HDI_index" ,
                       np.where(long_df['indicator'] == "31706", "Women_in_parliament_(%)" ,
                                long_df['indicator']))))
long_df
```

```python
# Concatenate the indicator and year columns to create the wide format column names for pivoting
df = long_df
df['indicator_year'] = df[['indicator', 'year']].agg('-'.join, axis=1)
df = df.drop(columns = ['indicator', 'year'])
df = df.rename(columns = {'country_code': 'ISO3'}, inplace = False)
df
```

```python
# Pivot df to receive the wide format results
df = df.pivot(index='ISO3', columns='indicator_year', values='value').reset_index()
df
```

```python
# Save the data to a csv
df.to_csv('HDRO_data.csv', index=False)
```
