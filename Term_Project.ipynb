```python
import pandas as pd
import requests
import json
```

### API - HDI data

```python
# Get request
resp = requests.get("http://ec2-54-174-131-205.compute-1.amazonaws.com/API/HDRO_API.php/indicator_id=137506/year=2019")
```

```python
# Load into json
hdi = json.loads(resp.text)
hdi
```

```python
# Check keys
hdi.keys()
```

```python
# indicator_value includes our observations
hdi['indicator_value']
```

```python
# Indicator value is a dictionary in which all country is a dictionary
hdi['indicator_value']['HUN']
```

```python
# the indicator and the year is a dictionary as well
hdi['indicator_value']['HUN']['137506']['2019']
```

```python
# Get country tickers
hdi['indicator_value'].keys()
```

```python
# Get ticker and HDI value
df_rows = []
for i in hdi.items():    
    for c in hdi['indicator_value'].keys():
        #print('country', c)
        print([c,hdi['indicator_value'][c]['137506']['2019']])
        df_rows.append([c,hdi['indicator_value'][c]['137506']['2019']])
df_rows
```

```python
type(df_rows)
```

```python
# Create dataframe
hdi_df = pd.DataFrame(df_rows, columns=['country_code','HDI_2019'])
```
