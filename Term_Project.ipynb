### API - HDI data

### Create long format
HDI Indicator and GDP per capita for 2012 and 2019

```python
# Get request
resp_all = requests.get("http://ec2-54-174-131-205.compute-1.amazonaws.com/API/HDRO_API.php/indicator_id=137506,194906/year=2016,2019")
```

```python
# Load into json
resp_all = json.loads(resp_all.text)
resp_all
```

```python
# Get all data
df_rows = []
for i in resp_all.items():    
    for c in resp_all['indicator_value'].keys():
        #print('country', c)
        for i in resp_all['indicator_value'][c].keys():
            #print('indicator', i)
            for y in resp_all['indicator_value'][c][i].keys():
                #print('year', y)
                print([c,resp_all['indicator_value'][c][i][y]])
                df_rows.append([c,i,y,resp_all['indicator_value'][c][i][y]])
df_rows
```

```python
# Create dataframe
hdi_df = pd.DataFrame(df_rows, columns=['country_code','indicator','year','value'])
hdi_df
```

```python
hdi_df.to_csv('long_format.csv', index=False)
```

### Load CPI data
